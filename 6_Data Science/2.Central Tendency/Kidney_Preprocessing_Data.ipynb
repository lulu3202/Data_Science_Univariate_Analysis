```python
import pandas as pd
```

```python
dataset=pd.read_csv("kidney_disease.csv")
```

```python
#1)Replace the missing value with central tendency-Mean Median Mode
#2)Entire Row delete
#3)REspective to the problem ,have to replace the missing values
#4)Consider this as a semi supervised, create a model, so that we can predict the missing values.
```

```python
# Drop the 'id' column as it is not useful for analysis
dataset.drop(columns=['id'], inplace=True)
```

```python
dataset
```

```python
# Filling missing values in categorical columns with the mode
for col in dataset.select_dtypes(include=['object']):
    dataset[col].fillna(dataset[col].mode()[0], inplace=True)
```

```python
dataset
```

```python
# Handling missing values by filling numerical columns with the mean
for col in dataset.select_dtypes(include=['float64']):
    dataset[col].fillna(dataset[col].mean(), inplace=True)
```
